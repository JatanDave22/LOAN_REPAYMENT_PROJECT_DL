```python
df["term"] = df["term"].apply(lambda x: int(x.split()[0]))
```

```python
df.head()
```

```python
df["home_ownership"].value_counts()
df["home_ownership"]=df["home_ownership"].replace(['NONE', 'ANY'], 'OTHER')
home_ownership_dummies=pd.get_dummies(df["home_ownership"],drop_first=True)
df.drop("home_ownership",axis=1,inplace=True)
df=pd.concat([df,home_ownership_dummies],axis=1)
df.head()
```

```python
zip_code=df["address"].apply(lambda x:x[-5:])
zip_code.value_counts()
zip_code_dummmies=pd.get_dummies(zip_code,drop_first=True)
zip_code_dummmies
df.drop("address",axis=1,inplace=True)
df=pd.concat([df,zip_code_dummmies],axis=1)
df.head()
```

```python
df["earliest_cr_line"]=df["earliest_cr_line"].apply(lambda x:int(x[-4:]))
df["earliest_cr_line"]
from datetime import datetime
now=datetime.now()
current_year=now.year
df["credit_history_years"]=df["earliest_cr_line"].apply(lambda x:current_year-x)
df["credit_history_years"]
df.drop("earliest_cr_line",axis=1,inplace=True)
```

```python
df["emp_length"] = df["emp_length"].replace({
    "< 1 year": "0",
    "10+ years": "10"
})

df["emp_length"] = df["emp_length"].str.replace(" years", "")
df["emp_length"] = df["emp_length"].str.replace(" year", "")

df["emp_length"] = df["emp_length"].astype(float)
median_emp_length = df["emp_length"].median()

df["emp_length"] = df["emp_length"].fillna(median_emp_length)
```

```python
df["loan_status"]=df["loan_status"].map({"Fully Paid":1,"Charged Off":0})
```

```python
df["verification_status"].value_counts()
verification_dummies=pd.get_dummies(df["verification_status"],drop_first=True)
df.drop("verification_status",axis=1,inplace=True)
df=pd.concat([df,verification_dummies],axis=1)
df.head()
```

```python
df = df.drop('title',axis=1)
```

```python
purpose_dummies = pd.get_dummies(df["purpose"], drop_first=True)

df.drop("purpose", axis=1, inplace=True)

df = pd.concat([df, purpose_dummies], axis=1)
```

```python
initial_dummies = pd.get_dummies(df["initial_list_status"], drop_first=True)

df.drop("initial_list_status", axis=1, inplace=True)

df = pd.concat([df, initial_dummies], axis=1)
```

```python
application_dummies = pd.get_dummies(df["application_type"], drop_first=True)

df.drop("application_type", axis=1, inplace=True)

df = pd.concat([df, application_dummies], axis=1)
```

```python
df["emp_title"].value_counts()
df.drop("emp_title", axis=1, inplace=True)
```

```python
df.drop("issue_d", axis=1, inplace=True)
```

```python
df["int_rate"].head()
```

```python
df["revol_util"]
df["revol_util"] = df["revol_util"].fillna(df["revol_util"].median())
```

```python
df["mort_acc"].isnull().sum()
mort_acc_avg = df.groupby("total_acc")["mort_acc"].mean()
mort_acc_avg
def fill_mort_acc(columns):

    mort_acc = columns["mort_acc"]
    total_acc = columns["total_acc"]

    if pd.isnull(mort_acc):
        return mort_acc_avg[total_acc] # total_acc becomes the INDEX of the Series
                                       # so mort_acc_avg[20] means:
                                       # "give average mort_acc for total_acc = 20"
    else:
        return mort_acc
df["mort_acc"] = df[["mort_acc", "total_acc"]].apply(fill_mort_acc, axis=1)
df["mort_acc"].isnull().sum()
```

```python
df["pub_rec_bankruptcies"].isnull().sum()
median_value = df["pub_rec_bankruptcies"].median()

df["pub_rec_bankruptcies"] = df["pub_rec_bankruptcies"].fillna(median_value)
df["pub_rec_bankruptcies"].isnull().sum()
```
